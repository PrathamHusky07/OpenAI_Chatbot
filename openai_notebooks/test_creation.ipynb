## Reducing the length of tokens by cutting the sentences

```python
import re
import pandas as pd
from typing import Set
from transformers import GPT2TokenizerFast
import numpy as np
from nltk.tokenize import sent_tokenize

tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

def count_tokens(text: str) -> int:
    """Count the number of tokens in a string"""
    return len(tokenizer.encode(text))

def reduce_long(
    long_text: str, long_text_tokens: bool = False, max_len: int = 590
) -> str:
    """
    Reduce a long text to a maximum of `max_len` tokens by potentially cutting at a sentence end
    """
    if not long_text_tokens:
        long_text_tokens = count_tokens(long_text)
    if long_text_tokens > max_len:
        sentences = sent_tokenize(long_text.replace("\n", " "))
        ntokens = 0
        for i, sentence in enumerate(sentences):
            ntokens += 1 + count_tokens(sentence)
            if ntokens > max_len:
                return ". ".join(sentences[:i]) + "."

    return long_text


# Specify the column with text data (replace 'text_column' with your column name)
df['content'] = df['content'].apply(lambda x: reduce_long(x, max_len=1000))  # Adjust max_len as needed

```

## Creating questions based on the context

```python
import openai
```

```python


api_key = "Your API Key"
def get_questions(context):
    try:
        response = openai.Completion.create(
            engine="davinci-instruct-beta-v3",
            prompt=f"Write questions based on the text below\n\nText: {context}\n\nQuestions:\n1.",
            temperature=1.0,
            max_tokens=257,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            stop=["\n\n"],
            api_key=api_key
        )
        return response['choices'][0]['text']
    except:
        return ""


df1['questions']= df1.context.apply(get_questions)
```

## Creating answers based on the context

```python
import openai  # Import the OpenAI package

api_key = "Your Api Key"

def get_answers(row):
    try:
        response = openai.Completion.create(
            engine="davinci-instruct-beta-v3",
            prompt=f"Write an answer based on the text below\n\nText: {row.context}\n\nQuestions:\n{row.questions}\n\nAnswers:\n1.",
            temperature=1.0,
            max_tokens=257,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            api_key=api_key
        )
        return response['choices'][0]['text']
    except Exception as e:
        print(e)
        return ""

# Assuming you have a DataFrame named df1
df1['answers'] = df1.apply(get_answers, axis=1)
df1['answers'] = "1. " + df1['answers']
df1 = df1.dropna().reset_index().drop('index', axis=1)

# Print the first answer
print(df1['answers'].values[0])
```
